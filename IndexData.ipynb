```python
# Importing pandas library for data manipulation
import pandas as pd
```

```python
df = pd.read_csv("imdb_top_1000.csv")
df.head()
```

```python
df.drop(columns=['Poster_Link', 'Meta_score', 'Gross', 'Star1', 'Star2', 'Star3' , 'Star4', 'No_of_Votes'], inplace=True)

df.head()
```

```python
#Checking if there are any missing values
df.isna().value_counts()
```

```python
# Filling missing values (NaN) in the DataFrame df with the string "None" and modifying the DataFrame in place
df.fillna("None", inplace=True)
```

```python
df.isna().value_counts()
```

```python
# Concatenate entire columns without spaces
df['search'] = df['Director'] + ' ' + df['Overview'] + ' ' + df['Series_Title']+ ' ' + df['Genre']

# Display the DataFrame with the new 'search' column
df.head()
```

```python
from sentence_transformers import SentenceTransformer # type: ignore

'''Choosing the 'all-mpnet-base-v2' model because it showcases the best average scores for both Performance Sentence Embeddings (14 Datasets)
and Performance Semantic Search (6 Datasets), making it a preferred choice for natural language processing tasks '''

model = SentenceTransformer("all-mpnet-base-v2")
```

```python
# Apply a lambda function to encode each overview text from the "Overview" column.
# Use the Sentence Transformer model stored in the variable model to encode the text.

df["SearchVector"] = df["search"].apply(lambda x: model.encode(x))
```

```python
df.head()
```

```python
es.ping()
```

```python
from IndexMapping import IndexMapping
```

```python
es.indices.create(index="search", mappings=IndexMapping)
```

```python
# The "to_dict" method with the argument "records" creates a dictionary for each row in the DataFrame.
record_list = df.to_dict("records")
```

```python
record_list[965]
```

```python
for record in record_list:
    try:
        # Check if 'Released_Year' value can be converted to numeric
        if isinstance(record['Released_Year'], (int, float)) or record['Released_Year'].isdigit():
            # Index the record only if it's numeric
            es.index(index="search", document=record)
        else:
            # Skip indexing if it's not numeric
            print(f"Skipping record with non-numeric 'Released_Year': {record}")
    except Exception as e:
        print(e)

```

```python
es.count(index="search")
```

```python
input_keyword = "christopher nolan"
input_representation = model.encode(input_keyword)

query = {
    "field" : "SearchVector",
    "query_vector" : input_representation,
    "k" : 3,
    "num_candidates" : 999, 
}

res = es.knn_search(index="search", knn=query , source=["Series_Title","Overview", "Genre", "Director"])
res["hits"]["hits"]
```
